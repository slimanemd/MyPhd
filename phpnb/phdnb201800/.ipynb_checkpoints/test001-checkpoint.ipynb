```python
N = 60000 # iterations
h = 0.0001
h2 = pow(h,2)
epsilon = 6.5 # n+1/2

y  = 0.0 
k = 0.0
x = -1*(N-2)*h

k_minus_2 = epsilon + x-2*h # k_0
k_minus_1 = epsilon + x-h # k_1
a = 0.1
y_minus_2 = 0 # y_0
y_minus_1 = a # y_1

x_out = []    ;y_out = []

n=-1*N+2
```

```python
while n<N-2:
  n+=1
  x += h;
  k = 2*epsilon - pow(x, 2)
  b = h2/12
  y = ( 2*(1-5*b*k_minus_1) * y_minus_1 - (1+b*k_minus_2) * y_minus_2 ) / (1 + b * k)

  # Save for plotting
  x_out.append(x)         ;y_out.append(y)

  # Shift for next iteration
  y_minus_2 = y_minus_1   ;y_minus_1 = y
  k_minus_2 = k_minus_1   ;k_minus_1 = k
```

```python
%matplotlib inline

#import numpy as np
import matplotlib.pyplot as plt

def doPlot(x_out0, y_out0, label0, title0):
    plt.figure(1)
    plt.plot(x_out0, y_out0, label=label0)  #"$\epsilon = "+repr(epsilon)+"$"
    plt.xlabel("x")   ;plt.ylabel("y")
    plt.title(title0)    #"dhgfhjsdgj hfjsdh")
    plt.legend()
    plt.show()
    
#doPlot(x_out, y_out, "yyyyy", "xxxxx")
```

```python
def doComputation():
    N, h, epsilon, a, k = 60000, 0.0001, 6.5, 0.1 , 0.0 # iterations

    h2  , n              = pow(h,2) , -1*N+2
    b                    = h2/12
    y, x                 = 0.0 , -1*(N-2)*h
    k_minus_2, k_minus_1 = epsilon + x-2*h , epsilon + x-h # k_1
    y_minus_2, y_minus_1 = 0 , a                          # y_0 # y_1
    x_out, y_out         = [], []
    while n<N-2:
        n +=1
        x += h;
        k  = 2*epsilon - pow(x, 2)

        y = ( 2*(1-5*b*k_minus_1) * y_minus_1 - (1+b*k_minus_2) * y_minus_2 ) / (1 + b * k)

        x_out.append(x)         ;y_out.append(y)   # Save for plotting
        y_minus_2 = y_minus_1   ;y_minus_1 = y     # Shift for next iteration
        k_minus_2 = k_minus_1   ;k_minus_1 = k
    return x_out, y_out
```

```python
#helper 01
func1 =  lambda x, y_1,  y_2 , k_1, k_2, b, k : \
    (2 * (1 - 5 * b * k_1) * y_1 - (1 + b * k_2) * y_2 ) / (1 + b * k)
#helper 02
func2 = lambda x, epsilon : 2 * epsilon - x**2   #pow(x, 2)

#task 01
def doComputation2(epsilon = 6.5):
    N, h, a, k  = 60000, 0.0001, 0.1 , 0.0        # parameters   iterations
    h2  , n, b           = h**2 , -1*N+2, h**2/12               #    b                    = h2/12
    
    y, x                 = 0.0 , -1*(N-2)*h
    k_2, k_1 = epsilon + x - 2 * h , epsilon + x - h      # k_1
    y_2, y_1 = 0 , a                                # y_0 # y_1
    x_out, y_out         = [], []
    while n < N-2: 
        n +=1
        x += h
        k  = func2(x,epsilon)
        y =  func1(x, y_1,  y_2 , k_1, k_2, b, k)

        x_out.append(x)         ;y_out.append(y)   # Save for plotting
        y_2, y_1 = y_1, y        # Shift for next iteration  #y_minus_2 = y_minus_1   ;y_minus_1 = y     # Shift for next iteration
        k_2, k_1 = k_1, k        #k_minus_2 = k_minus_1   ;k_minus_1 = k
    return x_out, y_out
```

```python
# computation
x_out, y_out  = doComputation2(epsilon)        #plt.plot(x_out, y_out, "$\epsilon = "+repr(epsilon)+"$","Schroedinger Eqn in Harmonic Potential")
doPlot(x_out, y_out, "$\epsilon = "+repr(epsilon)+"$","Schroedinger Eqn in Harmonic Potential")
```
